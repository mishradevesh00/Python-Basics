```python
data=('a,b,c,d\n'
     '1,2,3,4\n'
     '5,6,7,8\n'
     '9,10,11\n')
print(data)
```

```python
## giving data type manually using parameter dtype:

df2=pd.read_csv(StringIO(data),dtype=object)
df2
```

```python
df2=pd.read_csv(StringIO(data),dtype=int)
df2

## Error is bcs last value is not given yet, as far as object dtype is concerned it will take NaN
## But in case of Int or Float You have to give the Values.
```

```python
## Again defining the csv dataset by gining the last value also:

data=('a,b,c,d\n'
     '1,2,3,4\n'
     '5,6,7,8\n'
     '9,10,11,12\n')
print(data)
```

```python
df2=pd.read_csv(StringIO(data),dtype=int)
df2

## Now It will work w/o error.
```

```python
## Lets change the dtype as float
df2=pd.read_csv(StringIO(data),dtype=float)
df2
```

```python
## You can also give specific dtype to each column too:

df2=pd.read_csv(StringIO(data),dtype={'a':'Int64','b':float,'c':int})
df2
```

```python
## You can also access any value in this dataset by following way called as Indexing:

df2['b'][2]
```

```python
type(df2['b'][2])
```

```python
## To check data types of all the colums together:

df2.dtypes
```

```python
data=('indexx,a,b,c\n'
     '4,apple,bat,5.7\n'
     '8,orange,cow,10')
pd.read_csv(StringIO(data))
```

```python
## If I want to make my own index ie. say I want to make Indexx as my new index :

pd.read_csv(StringIO(data),index_col=0)
```

```python
## New dataset:

data=('a,b,c\n'
     '4,apple,bat,\n'
     '8,orange,cow,')
pd.read_csv(StringIO(data))
```

```python
## Use of index_col=False:

pd.read_csv(StringIO(data),index_col=False)
```

### Combining usecols & Index_col

```python
pd.read_csv(StringIO(data),usecols=['b','c'],index_col=False)
```

### Quoting and escape character:

```python
data='a,b\n"hello", \\"Bob\\", nice to meet you",5'
pd.read_csv(StringIO(data),escapechar='\\')
```

## *Reading Json to CSV :

```python
## Creating json dataset:

data='{"emp_name":"Devesh","email":"mishradvesh00@gmail.com","job_profile":[{"title1":"Team Lead","title2":"Developer"}]}'
df3=pd.read_json(data)
df3
```

```python
## Converting json to csv file

df3.to_json('EmpDetails.csv')
```

```python
pd.read_csv('EmpDetails.csv')
```

```python
df3.to_json()
```

```python
## Converting json to csv using orient parameter:

df3.to_json(orient="records")  ## It will give output record wise

## ie, emp_name:Devesh, email:mishradevesh00@gmail.com etc...
## while without orient=records, The above output is what we getting which is gicing o/p with row index associated 
# But using orient = records It is giving perfect key-value pair.
```

```python
data=('something,a,b,c,d,msg\n'
     'one,1,2,3.0,4\n'
     'two,5,6,,8,world\n'
     'three,9,10,11.0,12,foo')
dfs1=pd.read_csv(StringIO(data))
dfs1
```

```python
dfs1.to_csv('Test3.csv')
```

```python
pd.read_csv('Test3.csv')

## Open the file Test3.csv you can see the indexes and the header too.
```

```python
## In order to hide the indexes and the header:
dfs1.to_csv('Test3.csv',index=False,header=False)

## Now go check the file Test3.csv you will not see the header and the index of rows too.
```

```python
## What if I want only columns a,b,c to be there in the Test3.csv file:

dfs1.to_csv('Test3.csv',index=False,columns=['a','b','c'])
```

```python
## Now check the Test3.csv file you will have only these three columns a,b,c.
## or you can see the file here as:
pd.read_csv('Test3.csv')
```
