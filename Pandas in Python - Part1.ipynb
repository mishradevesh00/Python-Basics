### 3. By Changing the seperator, header. we can read values of file as:

```python
## Similarly we can read csv file using read_table, just by changing the seperator :

pd.read_table('testfile1.csv',sep=',')
```

```python
pd.read_table('testfile1.csv',sep=',',header=None)

## By default header always set to infer ie. first row is considered as column in csv file(see the o/p above)
## but if we set header = None then first row will not be considered as column, see below o/p:
```

```python
pd.read_table('testfile1.csv',sep=',',header='infer')
## It is bydefault header and by deafault o/p:
```

### 4. we can also change the header name as:


```python
pd.read_table('testfile1.csv',sep=',',header=None) ## first set header = none , now changing the names of header as:
```

```python
pd.read_table('testfile1.csv',sep=',',header=None,names=['col1','col2','col3','col4','col5'])
```

```python
pd.read_table('testfile1.csv',sep=',',header=None,names=['col1','col2','col3','col4','col5'],
             usecols=['col2','col3'])
```
