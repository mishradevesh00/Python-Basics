```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# global declaration on input x
x=np.linspace(-10,10,100)
```

#### 1. Sigmoid function

```python
def sigmoid(x):
    z=1/(1+np.exp(-x))
    return z
```

```python
plt.plot(x,sigmoid(x))
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.title("Sigmoid Function Plot")
plt.show()
```

#### 2.Binary Step Transfer Function

```python
def binaryStep(x):
    return np.heaviside(x,1)
```

```python
plt.plot(x,binaryStep(x))
plt.xlabel("x")
plt.ylabel("Binary Step")
plt.title('Binary Step Transfer Function')
plt.show
```

#### 3.Linear Transfer Function

```python
def linear(x):
    return x
```

```python
plt.plot(x,linear(x))
plt.xlabel('x')
plt.ylabel('linear')
plt.title('Linear Transfer Function')
plt.show
```

#### 4. softmax function

```python
def softmax(x):
    return ((np.exp(x))/np.sum(np.exp(x)))
```

```python
plt.plot(x,softmax(x))
plt.xlabel('x')
plt.ylabel('Softmax')
plt.title('Softmax Function Plot')
plt.show
```

#### 5.Tanh Transfer Function

```python
def tanh(x):
    return (1-np.exp(-2*x))/(1+np.exp(-2*x))
```

```python
plt.plot(x,tanh(x))
plt.xlabel('x')
plt.ylabel('Tanh')
plt.title('Tanh Transfer Function Plot')
plt.show()
```

#### 6. Rectified Linear unit transfer function(ReLu)

```python
def ReLu(x):
    input=[]
    for i in x:
        if i>0:
            input.append(i)
        else:
            input.append(0)
    return input
            
```

```python
plt.plot(x,ReLu(x))
plt.xlabel('x')
plt.ylabel('ReLu')
plt.title('Rectified Linear unit transfer function(ReLu)')
```
