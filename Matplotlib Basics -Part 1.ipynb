## MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

## Plotting using matplotlib
### 1.scatter plotting:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
x,y
```

### Creating Subplots:

1. In order to create a subplot we use plt.subplot()
2. Inside this subplot() we give various parameters.

```python
## Dividing canvas into 2*2 and plotting in each as per the intruction below:

plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'b--')

plt.subplot(2,2,3)
plt.plot(x,y,'g*--')

plt.subplot(2,2,4)
plt.plot(x,y,'y*--')
```

```python
## Suppose I want to plot in 1st and 4th subplot then:

plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,4)
plt.plot(x,y,'bo')
```

```python
## Subpart canvas into 2*3 (means 2 rows and 3 columns):

plt.subplot(2,3,1)
plt.plot(x,y,'r--')

plt.subplot(2,3,2)
plt.plot(x,y,'g--')

plt.subplot(2,3,3)
plt.plot(x,y,'go')

plt.subplot(2,3,4)
plt.plot(x,y,'bo')

plt.subplot(2,3,5)
plt.plot(x,y,'r--')

plt.subplot(2,3,6)
plt.plot(x,y,'b*--')
```

### Different types of curves created by plot:

```python
np.arange(0,4*np.pi,0.1)  ## range is 0 to 4*pi and step size of 0.1
```

```python
## We are applying sin ie. sin(0),sin(0.1).... and assigning it into y.
## Plotting sin curve:

x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)

plt.title("Sinewave Form")

plt.plot(x,y)
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, ie. (2*1 size 2 rows 1 column)
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
plt.plot(x, y_sin,'r--')  # Make the first plot 
plt.title('Sine')  

# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

### Bar Plot:

```python
x=[2,8,10]
y=[11,16,9]

x1=[3,9,11]
y1=[6,15,7]

plt.bar(x,y)
plt.bar(x1,y1,color='g')
plt.title("Bar Graph")
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

### Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

### Box Plot using matplotlib:

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);  
```

### Pie Chart:

```python

sizes=[25,25,25,25]
colors=['gold','yellow','red','green']

plt.pie(sizes,colors=colors)
```

```python
sizes=[30,30,20,20]
colors=['gold','yellow','red','green']

plt.pie(sizes,colors=colors)
```

```python
## Using explode parameters:

sizes=[25,25,25,25]
colors=['gold','yellow','red','green']

explode=(0,0,0,0)
plt.pie(sizes,colors=colors,explode=explode)
```

```python
## Using explode parameters:

sizes=[25,25,25,25]
colors=['gold','yellow','red','green']

explode=(0.2,0,0,0)
plt.pie(sizes,colors=colors,explode=explode)
```

```python
## Using labels:

## Using explode parameters:

sizes=[25,25,25,25]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
labels='C#','Python','Java','R'

explode=(0.2,0,0,0)
plt.pie(sizes,colors=colors,explode=explode,labels=labels)
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
